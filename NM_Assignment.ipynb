# Data visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
path="/content/drive/MyDrive/House Price India.csv"


house_data = pd.read_csv(path)
# print(house_data)
```

# Univariate Analysis

```python
plt.hist(house_data['living area'], bins=50)
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.title('Distribution of Living Area')
plt.show()
```

```python
print(house_data.describe())
```

# Bivariate Analysis

```python
x = house_data['living area']
y = house_data['Price']
plt.xlabel('Living Area')
plt.ylabel('Price')

plt.title("Relationship between Living Area and Price")
plt.plot(x,y)
plt.show()
```

# Multivariate Analysis

```python
pd.plotting.scatter_matrix(house_data[['Price', 'living area', 'lot area']], figsize=(10,10))
plt.show()

# Create box plots for selected variables
house_data[['Price', 'living area', 'lot area']].plot(kind='box', figsize=(10,5))
plt.show()
```
